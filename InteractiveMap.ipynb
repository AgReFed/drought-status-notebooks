```python
cities = pd.read_csv('Cities.csv')
cities.head(3)
```

```python
my_map = folium.Map(
         location = [13.1339, 16.1039],
         zoom_start = 2
)

for _, city in cities.iterrows():
    folium.Marker(
           location=[city['Latitude'], city['Longitude']],
           popup=city['Name'],
           tooltip=city['Name']
    ).add_to(my_map)

my_map
```

```python
def select_marker_color(row):
    if row['Lived'] == 'yes':
        return 'pink'
    elif row['Lived'] == 'no' and row['Visited'] == 'yes':
        return 'purple'
    return 'blue'
```

```python
cities['color'] = cities.apply(select_marker_color, axis=1)
cities.head(3)
```

```python
my_map = folium.Map(
         location = [13.1339, 16.1039],
         zoom_start = 2
)

for _, city in cities.iterrows():
    folium.Marker(
           location=[city['Latitude'], city['Longitude']],
           popup=city['Name'],
           tooltip=city['Name'],
           icon=folium.Icon(color=city['color'], prefix='fa', icon='circle')
    ).add_to(my_map)

my_map
```

```python
my_map.save('my_awesome_map.html')
```
